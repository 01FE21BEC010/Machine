even or odd

```python
num = int(input("Enter a number: "))
if (num % 2) == 0:
   print(" is Even")
else:
   print(" is Odd")
```

positive and negative

```python
num = (int(input("Enter a number: ")))
if num > 0:
  print("Positive number")
elif num == 0:
  print("Zero")
else:
  print("Negative number")
```

Grade

```python
score = 90
if score >= 90:
  print("You got an A.")
elif score >= 80:
   print("You got a B.")
elif score >= 70:
   print("You got a C.")
elif score >= 60:
  print("You got a D.")
else:
  print("You failed.")
```

Largest

```python
num1 = 10
num2 = 20
num3 = 30
if (num1 >= num2) and (num1 >= num3):
   largest = num1
elif (num2 >= num1) and (num2 >= num3):
   largest = num2
else:
   largest = num3

print("The largest number is", largest)
```
